```python
import numpy as np
import pandas as pd
import math

import matplotlib.pyplot as plt
```

```python
def string_to_int_mapping(strings):
    # Create a dictionary to store string to integer mapping
    string_to_int = {}
    int_sequence = []

    # Assign a unique integer to each unique string
    for string in strings:
        if string not in string_to_int:
            string_to_int[string] = len(string_to_int)  # Assign the next available integer
        # Append the corresponding integer to the sequence
        int_sequence.append(string_to_int[string])

    return int_sequence, string_to_int
```

```python
classes = np.load("Selected_Examples/y.npy").tolist()
```

```python
## First pair
file1 = "Selected_Examples/Layout1.npy"
scatter1 = np.load(file1)

df_scatter1 = pd.DataFrame(scatter1, columns=['x', 'y'])
df_scatter1["classes"] = classes
df_scatter1.to_csv("Selected_Examples/Layout1.csv")
```

```python
plt.figure(figsize=(8, 6))

# Use 'classes' to determine the color
scatter = plt.scatter(df_scatter1['x'], df_scatter1['y'], c=df_scatter1['classes'].astype('category').cat.codes, cmap='viridis')

# Add a colorbar
plt.colorbar(scatter, ticks=range(len(df_scatter1['classes'].unique())), label='Classes')

# Set labels and title
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Scatter Plot with Classes')

# Show the plot
plt.show()
```
