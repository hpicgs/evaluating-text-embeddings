```python
import pandas as pd
import numpy as np
```

```python
df_20newsgroup = pd.read_csv('results_cluster/cur_res/full_res_20_newsgroups.csv')
df_20newsgroup.head()
```

```python
df_emails = pd.read_csv('results_cluster/cur_res/full_res_emails.csv')
df_emails.head()
```

```python
df_bbc = pd.read_csv('results_cluster/cur_res/full_res_bbc_news.csv')
df_bbc.head()
```

```python
df_lyrics = pd.read_csv('results_cluster/cur_res/full_res_lyrics.csv')
df_lyrics.head()
```

```python
df_reuters = pd.read_csv('results_cluster/cur_res/full_res_reuters.csv')
df_reuters.head()
```

```python
df_7categories = pd.read_csv('results_cluster/cur_res/full_res_seven_categories.csv')
df_7categories.head()
```

```python
selected_cols = ['Trustworthiness', 'Continuity', 'Shephard Diagram Correlation', '7-Neighborhood Hit',
                'Distance consistency']


df_20newsgroup_short = df_20newsgroup.loc[:, selected_cols]
df_emails_short = df_emails.loc[:, selected_cols]
df_bbc_short = df_bbc.loc[:, selected_cols]
df_reuters_short = df_reuters.loc[:, selected_cols]
df_lyrics_short = df_lyrics.loc[:, selected_cols]
df_7categories_short = df_7categories.loc[:, selected_cols]
```

```python
concatenated_df = pd.concat([df_20newsgroup_short, df_emails_short, df_bbc_short, df_reuters_short, df_lyrics_short, df_7categories_short], axis=0)
```

```python
corr_matrix = concatenated_df.corr()
```

```python
corr_df_result = corr_matrix.stack().reset_index()
corr_df_result.columns = ['Variable 1', 'Variable 2', 'Correlation']
corr_df_result['Correlation'] = corr_df_result['Correlation'].round(decimals=2)

corr_df_result
```

```python
corr_df_result.to_csv("Results_Correlation_Final.csv")
```
